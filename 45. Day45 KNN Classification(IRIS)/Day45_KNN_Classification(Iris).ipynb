```python
#importing libraries
import numpy as np
import pandas as pd
```

**Data Pre-processing :**

```python
# Importing the dataset
df = pd.read_csv('iris.csv')
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
df['species'] = df['species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
df.head()
```

```python
df.tail()
```

```python
df=df.merge(pd.get_dummies(df.species, prefix='species'), left_index=True, right_index=True)
df.tail()
```

```python
df.merge(right=df, how='left', left_on='species', right_on='species')
```

```python
X = df.drop('species', axis=1)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 40)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Creating and Training KNN Model:**

```python
knn_clf = KNeighborsClassifier(n_neighbors=5, metric='euclidean', p=2)
knn_clf.fit(X_train, y_train)
```

**Predict Test Results:**

```python
y_pred = knn_clf.predict(X_test)
```

```python
pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

**Model Evaluation Metrics**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

score = accuracy_score(y_test, y_pred)
print(score)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
m1 = confusion_matrix(y_test, y_pred)
print(m1)
```
